```python
df1=pd.merge(cr_df,cj_df,how='outer')
df1
```

```python
df2=pd.merge(df1,products_df,how='outer')
df2
```

```python
df2.columns
```

```python
negative_review=df2[df2['Rating']<3].groupby(['ProductID','ProductName','Category','Stage']).mean(numeric_only=True).reset_index().sort_values(by='Rating',ascending=True)
negative_review
```

```python
# 1. Overall average rating (performance)
overall_rating = df2.groupby(['ProductID', 'ProductName', 'Category'])['Rating'].mean().reset_index()
overall_rating = overall_rating.rename(columns={'Rating': 'OverallRating'})

# 2. Negative review average
negative_review = df2[df2['Rating'] < 3] \
    .groupby(['ProductID', 'ProductName', 'Category'])['Rating'] \
    .mean().reset_index()
negative_review = negative_review.rename(columns={'Rating': 'AvgNegativeRating'})

# 3. % of negative reviews
negative_count = df2[df2['Rating'] < 3].groupby(['ProductID', 'ProductName', 'Category']).size().reset_index(name='NegativeCount')
total_count = df2.groupby(['ProductID', 'ProductName', 'Category']).size().reset_index(name='TotalCount')

# Merge and compute percentage
review_stats = total_count.merge(negative_count, on=['ProductID', 'ProductName', 'Category'], how='left')
review_stats['NegativeCount'] = review_stats['NegativeCount'].fillna(0)
review_stats['NegReviewPercent'] = (review_stats['NegativeCount'] / review_stats['TotalCount']) * 100

# 4. Combine all
product_analysis = overall_rating \
    .merge(negative_review, on=['ProductID', 'ProductName', 'Category'], how='left') \
    .merge(review_stats[['ProductID', 'ProductName', 'Category', 'NegReviewPercent']], on=['ProductID', 'ProductName', 'Category'], how='left')
```
